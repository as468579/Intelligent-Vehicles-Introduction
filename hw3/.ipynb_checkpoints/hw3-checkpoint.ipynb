```python
class ListNode():
    def __init__(self, val, next=None):
        self.val = val
        self.next = next
       self.visited = False
        
class Graphic():
    def __init__(self, filename):
        
        self.num_vertices = 0
        self.num_edges = 0
        self.adjlist = []
        
        try:
            f = open(filename)
            self.num_vertices = int(f.readline())
            self.num_edges = int(f.readline())
            self.adjlist = [None for _ in range(self.num_vertices)]

            for line in f.readlines():

                start, end = line.split()
                start, end = int(start), int(end)
                curr = None
                if self.adjlist[start]:
                    curr = self.adjlist[start]             

                self.adjlist[start] = ListNode(end, curr)

        except:
            print(f"Error loading file from {filename}")
            
    def dfs(self):
        for start_node in self.adjlist:
            self.dfs(start_node)
        
            
    def print_graph(self):
        print(f"Num of vertices : {self.num_vertices}")
        print(f"Num of edges : {self.num_edges}")
        print(f"Adjacency List : \n{self.adjlist}")
            
            
def checkCycle(filename):
    graph = Graphic(filename)
    graph.print_graph()
```

```python
checkCycle('Input1.dat')
```
